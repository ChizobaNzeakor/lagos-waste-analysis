```python
Kaivopuisto_polygon = gdf.geometry.iloc[0]
```

```python
buildings_gdf = ox.features_from_polygon(
    Kaivopuisto_polygon,
    tags = {"building": True, "highway": True, "railway": "station", "station": "subway"}
)
```

```python

     streets_graph = ox.graph_from_polygon(Kaivopuisto_polygon, network_type="all")
streets_gdf = ox.graph_to_gdfs(streets_graph, nodes=True)
```

```python
ax = buildings_gdf.plot(figsize=(50, 20), color="purple")
```
